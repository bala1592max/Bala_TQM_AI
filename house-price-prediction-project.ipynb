```python

import numpy as np 
import pandas as pd 
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df=pd.read_csv('/kaggle/input/competitions/home-data-for-ml-course/train.csv')
df.head()
```

```python
df.describe()
```

```python
null_cols=df.isnull().sum()
null_cols[null_cols>0]
```

```python
df=df.drop(columns=['Alley','PoolQC','Fence','MiscFeature'])
```

```python
null_cols=df.isnull().sum()
null_cols[null_cols>0]
```

```python
df['LotFrontage']=df['LotFrontage'].fillna(df['LotFrontage'].median())
df['FireplaceQu']=df['FireplaceQu'].fillna('None')
```

```python
null_cols=df.isnull().sum()
null_cols[null_cols>0]
```

```python
cols=['MasVnrArea','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','GarageType','GarageYrBlt','GarageFinish','GarageQual','GarageCond']
df[cols].dtypes
```

```python
cat_cols = ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Electrical','GarageType','GarageFinish','GarageQual','GarageCond']
for col in cat_cols:
    df[col] = df[col].fillna('None')
df['MasVnrArea']=df['MasVnrArea'].fillna(df['MasVnrArea'].median())
df['GarageYrBlt']=df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())
```

```python
null_cols=df.isnull().sum()
null_cols[null_cols>0]
```

```python
df.drop(columns=['BsmtQual'])
```

```python
cat_cols=df.select_dtypes(include=['object']).columns
cat_cols
```

```python
df_encoded=pd.get_dummies(df,columns=cat_cols)
df_encoded
```

```python
df_encoded['KitchenQual_Ex'].unique()
```

```python
bool_cols=df.select_dtypes(include=['bool']).columns
bool_cols
```

```python
df_encoded.dtypes
```

```python
# Calculate the correlation for all numerical columns
corr_matrix = df_encoded.corr()

# Isolate SalePrice and sort to find the strongest predictors
print(corr_matrix['SalePrice'].sort_values(ascending=False).head(15))
```

```python
df_test=pd.read_csv('/kaggle/input/competitions/home-data-for-ml-course/test.csv')
df_test
```

```python
features=['OverallQual','GrLivArea','GarageArea','TotalBsmtSF',
'1stFlrSF']
x=df_encoded[features]
y=df_encoded['SalePrice']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred_train=model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
rmse=np.sqrt(mean_squared_error(y_train,y_pred_train))
r2=r2_score(y_train,y_pred_train)
print(f"RMSE train: {rmse:,.2f}")
print(f"R2 train: {r2:,.2f}")
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
r2=r2_score(y_test,y_pred)
print(f"RMSE test: {rmse:,.2f}")
print(f"R2 test: {r2:,.2f}")
```
